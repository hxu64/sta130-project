### Q1: How does the frequency of social interaction with family, friends, and neighbors relate to feelings of existential loneliness and emotional well-being among survey respondents?

This questions explores how the frequency of social interaction with family, friends, and neighbors relates to existential loneliness and emotional well-being. The goal is to determine if more frequent social interactions are linked to lower loneliness and higher well-being scores.

1. **Social Interaction Frequency**  
   - **Variables:** `CONNECTION_social_days_family_p7d`, `CONNECTION_social_days_friends_p7d`, `CONNECTION_social_days_neighbours_p7d`
   - **Response Categories:** none, some (1-3), most (4-6), every day
   - **Visualization:** Bar charts will display the distribution of responses across different interaction frequencies.

2. **Loneliness Metrics**  
   - **Variables:** `LONELY_existential_loneliness_scale_outlook`, `LONELY_existential_loneliness_scale_share`
   - **Response Scale:** 1-9 (1 = strongly agree, 9 = strongly disagree) 
   - **Visualization:** Box plots will illustrate loneliness scores grouped by interaction frequency.

3. **Emotional Well-Being**  
   - **Variable:** `LONELY_dejong_emotional_social_loneliness_scale_emptiness`
   - **Response Categories:** yes, no, more or less 
   - **Visualization:** Stacked bar charts will compare well-being states across interaction levels.

**Analysis Plan:**
1. Calculate descriptive statistics for each variable to summarize mean and standard deviation for loneliness and well-being scores at each interaction frequency level.
2. Use bootstrapping to estimate confidence intervals for mean loneliness and emotional well-being scores. Non-overlapping confidence intervals will suggest significant differences across interaction levels.
3. Set up linear regression models with loneliness and emotional well-being as outcomes to assess the impact of social interaction frequency on these variables.

**Assumptions** 
-  loneliness scale has a linear relationship with the frequency of social interactions.
- Bootstrapping assumes that our sample is representative of the population and is independent.

**Hypotheses:**
- **Existential Loneliness:** H<sub>0</sub>: No association between social interaction frequency and loneliness scores. We expect a negative association; more frequent interactions should correlate with lower loneliness.
- **Emotional Well-Being:** H<sub>0</sub>: No association between social interaction frequency and well-being scores. We anticipate a positive association; more social interactions should relate to higher well-being.

**Expected Results:**
- **Bootstrapping:** Expect narrower, lower confidence intervals for loneliness scores at higher interaction frequencies, with significant differences indicated by non-overlapping intervals.
- **Linear Regression:** Anticipate negative coefficients for loneliness scores and positive coefficients for emotional well-being at higher interaction levels compared to the "rarely" baseline.

**Relevance:** If confirmed, these findings suggest that increased social interaction may protect against loneliness and enhance emotional well-being, informing mental health interventions and public health initiatives aimed at reducing social isolation.

### Q2: Is there an association between self-perceived mental health, self-esteem, and relationship satisfaction, and how does loneliness influence these associations?

This question investigates the relationships between self-perceived mental health, self-esteem, and relationship satisfaction, focusing on the moderating role of loneliness. We aim to understand if individuals with higher loneliness scores report lower self-esteem, mental health, and relationship satisfaction by analyzing mean scores across varying levels of loneliness.

1. **Self-Perceived Mental Health**  
   - **Variables**: `WELLNESS_self_rated_mental_health`, `WELLNESS_subjective_happiness_scale_happy`, `PSYCH_self_esteem_unknown_scale_think_of_me`  
   - **Purpose**: Assess self-assessments on mental health and happiness to determine if higher ratings correlate with higher self-esteem and relationship satisfaction.  
   - **Visualization**: Histograms for distributions and scatter plots to identify associations between happiness and mental health variables.

2. **Self-Esteem Metrics**  
   - **Variables**: `PSYCH_rosenberg_self_esteem_satisfied`, `PSYCH_rosenberg_self_esteem_worth`, `PSYCH_rosenberg_self_esteem_good_qualities`  
   - **Purpose**: Explore correlations between self-esteem and relationship satisfaction.  
   - **Visualization**: Stacked bar charts to compare self-esteem distributions across mental health levels.

3. **Social Satisfaction**  
   - **Variable**: `WELLNESS_satisfied_relationship`  
   - **Purpose**: Investigate the relationship between social satisfaction and mental health, assessing if stronger ties improve well-being.  
   - **Visualization**: Violin plots to show satisfaction distributions across self-esteem and mental health levels.

4. **Loneliness Metric**  
   - **Variable**: `LONELY_existential_loneliness_scale_understand`  
   - **Purpose**: Examine if high loneliness scores are linked to lower mental health and relationship satisfaction.  
   - **Visualization**: Bar plots to categorize loneliness levels and average relationship satisfaction scores.

#### Analysis Plan

1. **Data Summary with pandas**: Calculate summary statistics for each variable and visualize patterns with grouped bar plots to identify trends.

2. **Bootstrapped Confidence Intervals**: Construct confidence intervals around mean scores for self-esteem and relationship satisfaction across mental health categories to evaluate significant differences.

3. **Permutation Test**: Assess the statistical significance of associations between self-perceived mental health and self-esteem, and similarly for mental health and relationship satisfaction.

4. **Linear Regression**: Use regression with indicator variables for mental health levels to quantify the effects on self-esteem and relationship satisfaction.

**Assumption**
- the samples drawn from the population are independent of one another, allowing for valid comparisons between groups and ensuring that results from one individual do not influence those of another.
- the scales used to measure self-perceived mental health, self-esteem, and relationship satisfaction are reliable, meaning they consistently yield the same results under similar conditions, ensuring that the assessments accurately reflect the constructs of interest.
- there are linear relationships between self-perceived mental health, self-esteem, and relationship satisfaction, allowing for the application of linear regression techniques to analyze the data effectively.

#### Hypothesis and Expected Results

1. **Self-Esteem and Mental Health**  
   - **Null Hypothesis (H₀)**: No significant differences in self-esteem across mental health categories.  
   - **Alternative Hypothesis (Hₐ)**: Higher mental health ratings correlate with higher self-esteem.  
   - **Expected Result**: Significant differences in bootstrapped confidence intervals and positive regression coefficients confirming the association.

2. **Relationship Satisfaction and Mental Health**  
   - **Null Hypothesis (H₀)**: No significant differences in relationship satisfaction across mental health levels.  
   - **Alternative Hypothesis (Hₐ)**: Higher mental health ratings are linked to greater relationship satisfaction.  
   - **Expected Result**: Positive findings in bootstrapped intervals, significant permutation test results, and positive regression coefficients indicating a strong association.

   - **Visualization**: A bar chart displaying engagement frequency will illustrate the relationship with loneliness.

2. **Loneliness**  
   - **Variable**: `LONELY_ucla_loneliness_scale_left_out`  
   - **Categories**: Hardly, Sometimes, Often  
   - **Purpose**: Assesses how often respondents feel lonely; higher loneliness is expected among those who rarely exercise socially.  
   - **Visualization**: A grouped bar chart comparing loneliness levels by exercise engagement frequency.

3. **Relational Satisfaction**  
   - **Variable**: `PSYCH_relational_needs_satisfaction_scale_7pt_true_self_without_rejection`  
   - **Scale**: 1-9 (1 is strongly agree, 9 is strongly disagree)  
   - **Purpose**: Gauges comfort in showing one’s true self; aims to correlate with social exercise settings.  
   - **Visualization**: A scatter plot of relational satisfaction scores by exercise frequency.

#### Analysis Plan

1. **Descriptive Statistics**: Calculate averages for mental health, self-esteem, and relational satisfaction across social media usage levels, visualized with bar or box plots to identify trends.
  
2. **Bootstrapped Confidence Intervals**: Create confidence intervals for mental health, self-esteem, and relational satisfaction to compare average scores between different social media usage groups.

3. **Permutation Tests**: Conduct permutation tests to determine if observed differences in scores are statistically meaningful, comparing high vs. low social media usage.

4. **Linear Regression**: Model relationships using social media usage levels as an indicator variable to quantify effects on mental health and relational outcomes.

**Assumption**
- the responses from participants regarding their social engagement during physical activities are independent, meaning that one participant's level of engagement does not influence another's

- variances of loneliness and self-expression scores are consistent across different levels of social engagement, which is important for the validity of the statistical tests employed, such as linear regression and permutation tests.

#### Hypotheses

1. **Loneliness**  
   - **Null Hypothesis (H<sub>0</sub>)**: No association between social engagement frequency and loneliness levels.

2. **Self-Expression**  
   - **Null Hypothesis (H<sub>0</sub>)**: No association between social engagement frequency and the ability to express one’s true self.

#### Expected Results

1. **Loneliness**: Anticipate lower loneliness scores among those with higher social engagement, with bootstrapped confidence intervals indicating significant differences and regression analysis showing negative coefficients for higher engagement levels.
